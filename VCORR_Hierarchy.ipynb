```python
import pandas as pd
import scipy.cluster.hierarchy as spc
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
from matplotlib.mlab import PCA

%matplotlib notebook
```

```python
df = pd.read_csv('load_voltage_profile_55.csv')
df['timestamp']=pd.to_datetime(df['timestamp'])
df=df.set_index('timestamp').sort_index().interpolate().ffill().bfill()
df.head()
```

```python
corr = df.corr()#.values()
corr.head()
```

```python
pdist = spc.distance.pdist(corr)
linkage = spc.linkage(pdist, method='complete')
idx = spc.fcluster(linkage, 0.5 * pdist.max(), 'distance')
```

```python
Phase_X,Phase_Y,Phase_Z=[],[],[]

[Phase_X.append('Load'+str(load_number+1))  for (load_number,cluster) in enumerate (idx) if cluster==1]
[Phase_Y.append('Load'+str(load_number+1))  for (load_number,cluster) in enumerate (idx) if cluster==2]
[Phase_Z.append('Load'+str(load_number+1))  for (load_number,cluster) in enumerate (idx) if cluster==3]

phaseX_df = pd.DataFrame(Phase_X,columns =['Phase X']) 
phaseY_df = pd.DataFrame(Phase_Y,columns =['Phase Y']) 
phaseZ_df = pd.DataFrame(Phase_Z,columns =['Phase Z']) 
```

```python
phase_Identification_df = pd.concat([phaseX_df, phaseY_df, phaseZ_df], axis=1, ignore_index=True, sort=False)
phase_Identification_df.columns = ['phase X', 'phase Y', 'phase Z']
phase_Identification_df.index.name = 'Index'
phase_Identification_df
```

```python
phase_Identification_df.to_csv('phase_Identification.csv')
```
