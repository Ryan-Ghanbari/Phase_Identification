```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np

%matplotlib notebook
```

```python
df = pd.read_csv('load_voltage_profile_55.csv')
df['timestamp']=pd.to_datetime(df['timestamp'])
df=df.set_index('timestamp').sort_index().interpolate().ffill().bfill()
df.head()
```

```python
df_corr = df.corr()
df_corr.head()
```

```python
df_corr_inv = pd.DataFrame(np.reciprocal(df_corr.values), df_corr.columns, df_corr.index)
df_corr_inv.head()
```

```python
df_corr_85 = df_corr[df_corr>0.9].stack()
df_corr_85
```

```python
phase_x_anchor =  df_corr_85.reset_index().groupby('level_0').count().sort_values('level_1', ascending=False).index[0]
phase_x = df_corr_85[df_corr_85.index.get_level_values(0)==phase_x_anchor].index.get_level_values(1).tolist()
phase_y_anchor = df_corr[df_corr.index==phase_x_anchor].idxmin(axis=1).values[0]
phase_y = df_corr_85[df_corr_85.index.get_level_values(0)==phase_y_anchor].index.get_level_values(1).tolist()

potential_z = [vol_l for vol_l in df_corr.columns if vol_l not in phase_x and vol_l not in phase_y]



def get_phase_z_anchor(x_a, y_a):
    distances = []
    for vol_l in potential_z:
        x_cor = df_corr[df_corr.index==x_a][vol_l].values[0]
        y_cor = df_corr[df_corr.index==y_a][vol_l].values[0]
        dist = (1-x_cor) + (1-y_cor)
        distances.append(dist)
    return potential_z[distances.index(min(distances))]

phase_z_anchor = get_phase_z_anchor(phase_x_anchor,phase_y_anchor)
phase_z = df_corr_85[df_corr_85.index.get_level_values(0)==phase_z_anchor].index.get_level_values(1).tolist()

len(phase_x), len(phase_y), len(phase_z)
```

## Dealing with common members and remaining points

```python
def intersection(lst1, lst2):  
    temp = set(lst2) 
    lst3 = [value for value in lst1 if value in temp] 
    return lst3

def get_common_members(phase_x, phase_y, phase_z):
    xy = intersection(phase_x, phase_y)
    yz = intersection(phase_y, phase_z)
    zx = intersection(phase_z, phase_x)
    return list(set(xy+ yz+ zx))

def get_remaining_points(phase_x, phase_y, phase_z):
    return [pn for pn in df.columns if pn not in phase_x + phase_y + phase_z]
    

commons = get_common_members(phase_x, phase_y, phase_z)
remainings = get_remaining_points(phase_x, phase_y, phase_z)
len(remainings), len(commons)
```

```python
def get_cor_distance(p1, p2):
    p_cor = df_corr[df_corr.index==p1][p2].values[0]
    p_dist = 1-p_cor
    return p_dist

def get_list_distance(p1, lst1):
    dist = 0
    for p2 in lst1:
        dist+=get_cor_distance(p1, p2)
    return dist/len(lst1)

#get_cor_distance(commons[0],phase_z_anchor)    
```

```python
def difference(List_A, List_B):
    #List_A, List_B,dif = set(List_A), set(List_B),[]
    dif=[]
    [dif.append(item) for item in set(List_A) if item not in set(List_B)]
    return dif
 
phase_x_polished=difference(phase_x,commons)
phase_y_polished=difference(phase_y,commons)
phase_z_polished=difference(phase_z,commons)
```

```python
phase_list_polished = [phase_x_polished, phase_y_polished, phase_z_polished]
def add_extras_2_closest_phase(p1, phase_list_polished):
    phase_distances = []
    for phase_n_polished in phase_list_polished:
        dist = get_list_distance(p1, phase_n_polished)
        phase_distances.append(dist)
    min_phase = np.argmin(phase_distances)
    phase_list_polished[min_phase].append(p1)
    return [list(set(lst)) for lst in phase_list_polished]
    
    
for p1 in commons: 
    for phase_list in phase_list_polished:
        print(p1, len(phase_list))
    phase_list_polished = add_extras_2_closest_phase(p1, phase_list_polished)
    for phase_list in phase_list_polished:
        print(p1, len(phase_list))
print('-------------')
for p1 in remainings: 
    for phase_list in phase_list_polished:
        print(p1, len(phase_list))
    phase_list_polished = add_extras_2_closest_phase(p1, phase_list_polished)
    for phase_list in phase_list_polished:
        print(p1, len(phase_list))

for phase_list in phase_list_polished:
    print(len(phase_list))
```
